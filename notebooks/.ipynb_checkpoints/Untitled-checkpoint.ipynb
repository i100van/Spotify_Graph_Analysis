```python
import pandas as pd
import numpy as np
import ast
```

```python
df=pd.read_csv('../data/data.csv')
def is_duo(x):
    arts=ast.literal_eval(x)
    if len(arts)>1:
        return True
    else:
        return False
    
df['is_colab']=df['artists'].apply(lambda x:is_duo(x))
df=df[df.is_colab==True]
```
