```python
import pandas as pd
import numpy as np
import ast
import networkx as nx
import matplotlib.pyplot as plt
from pyvis.network import Network
```

```python
df=pd.read_csv('../data/data.csv')
def is_duo(x):
    arts=ast.literal_eval(x)
    if len(arts)>1:
        return True
    else:
        return False
    
df['is_colab']=df['artists'].apply(lambda x:is_duo(x))
df=df[(df.is_colab==True) & (df.popularity>50)]
```

```python
def add_node(x,graph):
    arts=ast.literal_eval(x)
    for art in arts:
        graph.add_node(art)
    for i in range(len(arts)-1):
        graph.add_edge(arts[i],arts[i+1])

G = nx.Graph()
df['artists'].apply(lambda x:add_node(x,G))
```

```python
G.number_of_nodes(),G.number_of_edges()
```

```python
g = Network()
g.from_nx(G)
g.show_buttons(filter_=['physics'])
g.show("nx.html")
```
